# Lebel Encoding with loop

```python
df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df2.sex=le.fit_transform(df2.sex)
df2.smoker=le.fit_transform(df2.smoker)
df2.region=le.fit_transform(df2.region)
df2.head()
```

```python
df2.columns
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in df2.columns:
    if is_numeric_dtype(df2[col]):
        continue
    else:
        df2[col]=le.fit_transform(df2[col])
```

```python
df2.head()
```

# One hot encoding

```python
df3.head()
```

```python
for col in df3.columns:
    if is_numeric_dtype(df3[col]):
        continue
    else:
        dum = pd.get_dummies(df3[col],prefix=col,drop_first=True)
        df3=pd.concat([df3,dum],axis=1)
        df3=df3.drop(col,axis=1)
```

```python
df3.head()
```

# Ordinal

```python
df4.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
var=df4.sex.unique()
ordinal=OrdinalEncoder(categories=[var])
encoded=ordinal.fit_transform(df4[['sex']])  
en=pd.DataFrame(encoded,columns=['SEX']) 
df4=df4.drop('sex',axis=1)  
df4=pd.concat([df4,en],axis=1) 
```

```python
var=df4.smoker.unique()
ordinal=OrdinalEncoder(categories=[var])
encoded=ordinal.fit_transform(df4[['smoker']])  
en=pd.DataFrame(encoded,columns=['SMOKER']) 
df4=df4.drop('smoker',axis=1)  
df4=pd.concat([df4,en],axis=1) 
```

```python
var=df4.region.unique()
ordinal=OrdinalEncoder(categories=[var])
encoded=ordinal.fit_transform(df4[['region']])  
en=pd.DataFrame(encoded,columns=['REGION']) 
df4=df4.drop('region',axis=1)  
df4=pd.concat([df4,en],axis=1) 
```
